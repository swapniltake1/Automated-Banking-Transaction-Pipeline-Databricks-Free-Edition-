```python
# ============================================================
# Banking Data Pipeline
# Notebook : 00_Precheck
# Purpose  : Validate Project Folder Structure
# ============================================================

BASE_PATH = "/Volumes/banking_catalog/banking_project/banking_volume"

print("=" * 80)
print("Starting Precheck Validation")
print("=" * 80)
```

```python
# ============================================================
# Validate Main Project Folders
# ============================================================

REQUIRED_FOLDERS = [
    "incoming",
    "processed",
    "rejected",
    "archive",
    "bronze",
    "silver",
    "gold",
    "checkpoints",
    "logs"
]

missing_folders = []

print("\nChecking Main Project Folders...\n")

for folder in REQUIRED_FOLDERS:

    folder_path = f"{BASE_PATH}/{folder}"

    try:

        dbutils.fs.ls(folder_path)

        print(f"✅ PASS : {folder}")

    except Exception:

        print(f"❌ FAIL : {folder}")

        missing_folders.append(folder)

if missing_folders:

    raise Exception(
        f"Missing Project Folders : {', '.join(missing_folders)}"
    )

print("\n✅ Project Folder Validation Completed Successfully")
```

```python
# ============================================================
# Discover Source Systems
# ============================================================

SOURCE_SYSTEMS = sorted([
    folder.name.replace("/", "")
    for folder in dbutils.fs.ls(f"{BASE_PATH}/incoming")
])

print("\nSource Systems Found\n")

for source in SOURCE_SYSTEMS:
    print(source)
```

```python
# ============================================================
# Validate Source Folder Structure
# ============================================================

LANDING_FOLDERS = [
    "incoming",
    "processed",
    "rejected",
    "archive"
]

missing_paths = []

print("\nChecking Source Folder Structure...\n")

for landing in LANDING_FOLDERS:

    for source in SOURCE_SYSTEMS:

        folder_path = f"{BASE_PATH}/{landing}/{source}"

        try:

            dbutils.fs.ls(folder_path)

            print(f"✅ PASS : {landing}/{source}")

        except Exception:

            print(f"❌ FAIL : {landing}/{source}")

            missing_paths.append(f"{landing}/{source}")

if missing_paths:

    raise Exception(
        "Missing Source Folders:\n" +
        "\n".join(missing_paths)
    )

print("\n✅ Source Folder Validation Completed Successfully")
```

```python
# ============================================================
# Validate Medallion Layer
# ============================================================

LAYERS = [
    "bronze",
    "silver",
    "gold"
]

missing_layers = []

print("\nChecking Medallion Layers...\n")

for layer in LAYERS:

    path = f"{BASE_PATH}/{layer}"

    try:

        dbutils.fs.ls(path)

        print(f"✅ PASS : {layer}")

    except Exception:

        print(f"❌ FAIL : {layer}")

        missing_layers.append(layer)

if missing_layers:

    raise Exception(
        f"Missing Layers : {', '.join(missing_layers)}"
    )

print("\n✅ Medallion Layer Validation Completed")
```

```python
# ============================================================
# Read Access Validation
# ============================================================

print("\nChecking Read Access...\n")

try:

    display(dbutils.fs.ls(BASE_PATH))

    print("✅ Read Access Available")

except Exception as ex:

    raise Exception(
        f"Unable to access project folders.\n{str(ex)}"
    )
```

```python
# ============================================================
# Precheck Summary
# ============================================================

print("=" * 80)
print("BANKING DATA PIPELINE PRECHECK COMPLETED")
print("=" * 80)

print(f"Project Path      : {BASE_PATH}")
print(f"Source Systems    : {len(SOURCE_SYSTEMS)}")
print(f"Landing Folders   : 4")
print(f"Pipeline Status   : READY")

print("=" * 80)
print("✅ All validations completed successfully.")
print("=" * 80)
```
