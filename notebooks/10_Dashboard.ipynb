```python
from pyspark.sql.functions import *

pipeline_start("10_Dashboard")
```

```python
daily = spark.table(f"{CATALOG_NAME}.{SCHEMA_NAME}.gold_daily_summary")
branch = spark.table(f"{CATALOG_NAME}.{SCHEMA_NAME}.gold_branch_summary")
customer = spark.table(f"{CATALOG_NAME}.{SCHEMA_NAME}.gold_customer_summary")
fraud = spark.table(f"{CATALOG_NAME}.{SCHEMA_NAME}.gold_fraud_transactions")

display(daily)
display(branch)
display(customer)
display(fraud)
```

```python
spark.sql(f"""
SELECT
COUNT(*) AS Total_Days,
SUM(total_transactions) AS Total_Transactions,
ROUND(SUM(total_amount),2) AS Total_Amount
FROM {CATALOG_NAME}.{SCHEMA_NAME}.gold_daily_summary
""")
```

```python
display(
spark.sql(f"""
SELECT
transaction_date,
total_transactions,
total_amount
FROM {CATALOG_NAME}.{SCHEMA_NAME}.gold_daily_summary
ORDER BY transaction_date
""")
)
```

```python

display(
spark.sql(f"""
SELECT
branch,
total_transactions,
total_amount
FROM {CATALOG_NAME}.{SCHEMA_NAME}.gold_branch_summary
ORDER BY total_amount DESC
""")
)
```

```python
display(
spark.sql(f"""
SELECT
customer_id,
transaction_count,
total_amount,
highest_transaction
FROM {CATALOG_NAME}.{SCHEMA_NAME}.gold_customer_summary
ORDER BY total_amount DESC
LIMIT 20
""")
)
```

```python
display(
spark.sql(f"""
SELECT
fraud_reason,
COUNT(*) AS fraud_count
FROM {CATALOG_NAME}.{SCHEMA_NAME}.gold_fraud_transactions
GROUP BY fraud_reason
ORDER BY fraud_count DESC
""")
)
```
