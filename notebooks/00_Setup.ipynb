```python
%sql
CREATE CATALOG IF NOT EXISTS banking_catalog;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS banking_catalog.banking_project;
```

```python
%sql
CREATE VOLUME IF NOT EXISTS banking_catalog.banking_project.banking_volume;
```

```python
# ============================================================
# Banking Data Pipeline
# Notebook : 00_Setup_Folders
# Purpose  : Create Project Folder Structure
# ============================================================

BASE_PATH = "/Volumes/banking_catalog/banking_project/banking_volume"

SOURCE_SYSTEMS = [
    "ATM",
    "UPI",
    "NEFT",
    "RTGS",
    "CARD",
    "INTERNET_BANKING",
    "MOBILE_BANKING",
    "BRANCH"
]

print("=" * 80)
print("Starting Banking Project Setup")
print("=" * 80)
```

```python
# ============================================================
# Create Main Project Folders
# ============================================================

PROJECT_FOLDERS = [

    # Landing Zone
    "incoming",
    "processed",
    "rejected",
    "archive",

    # Medallion Layers
    "bronze",
    "silver",
    "gold",

    # Pipeline Support
    "checkpoints",
    "logs",
    "metadata",
    "temp"
]

print("\nCreating Main Project Folders...\n")

for folder in PROJECT_FOLDERS:

    folder_path = f"{BASE_PATH}/{folder}"

    dbutils.fs.mkdirs(folder_path)

    print(f"✅ {folder}")
```

```python
# ============================================================
# Create Source System Folders
# ============================================================

LANDING_FOLDERS = [
    "incoming",
    "processed",
    "rejected",
    "archive"
]

print("\nCreating Banking Source Folders...\n")

for landing in LANDING_FOLDERS:

    for source in SOURCE_SYSTEMS:

        dbutils.fs.mkdirs(f"{BASE_PATH}/{landing}/{source}")

        print(f"✅ {landing}/{source}")
```

```python
# ============================================================
# Create Medallion Layer Structure
# ============================================================

MEDALLION_STRUCTURE = {

    "bronze": [
        "transactions",
        "customers",
        "accounts",
        "branches"
    ],

    "silver": [
        "transactions",
        "customers",
        "accounts",
        "branches"
    ],

    "gold": [
        "daily_summary",
        "customer_summary",
        "branch_summary",
        "fraud_transactions"
    ]

}

print("\nCreating Medallion Layer Structure...\n")

for layer, folders in MEDALLION_STRUCTURE.items():

    for folder in folders:

        dbutils.fs.mkdirs(f"{BASE_PATH}/{layer}/{folder}")

        print(f"✅ {layer}/{folder}")
```

```python
# ============================================================
# Create Checkpoint Folders
# ============================================================

CHECKPOINT_FOLDERS = [

    "transactions",
    "customers",
    "accounts",
    "branches",
    "fraud_detection"

]

print("\nCreating Checkpoint Folders...\n")

for folder in CHECKPOINT_FOLDERS:

    dbutils.fs.mkdirs(f"{BASE_PATH}/checkpoints/{folder}")

    print(f"✅ checkpoints/{folder}")
```

```python
# ============================================================
# Create Log Folder Structure
# ============================================================

LOG_FOLDERS = [

    "pipeline_logs",
    "error_logs",
    "audit_logs"

]

print("\nCreating Log Folders...\n")

for folder in LOG_FOLDERS:

    dbutils.fs.mkdirs(f"{BASE_PATH}/logs/{folder}")

    print(f"✅ logs/{folder}")
```

```python
# ============================================================
# Verify Project Structure
# ============================================================

print("=" * 80)
print("Project Root")
print("=" * 80)

display(dbutils.fs.ls(BASE_PATH))

print("=" * 80)
print("Incoming Source Systems")
print("=" * 80)

display(dbutils.fs.ls(f"{BASE_PATH}/incoming"))

print("=" * 80)
print("Bronze Layer")
print("=" * 80)

display(dbutils.fs.ls(f"{BASE_PATH}/bronze"))
```

```python
# ============================================================
# Setup Summary
# ============================================================

print("=" * 80)
print("BANKING PROJECT SETUP COMPLETED")
print("=" * 80)

print(f"Project Path       : {BASE_PATH}")
print(f"Source Systems     : {len(SOURCE_SYSTEMS)}")
print(f"Landing Zones      : {len(LANDING_FOLDERS)}")
print(f"Medallion Layers   : {len(MEDALLION_STRUCTURE)}")
print(f"Checkpoint Folders : {len(CHECKPOINT_FOLDERS)}")

print("=" * 80)
print("✅ Folder structure created successfully.")
print("=" * 80)
```
