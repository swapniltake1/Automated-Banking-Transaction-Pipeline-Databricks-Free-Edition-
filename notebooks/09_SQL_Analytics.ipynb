```python
spark.sql(f"""
CREATE OR REPLACE VIEW {CATALOG_NAME}.{SCHEMA_NAME}.vw_daily_summary AS
SELECT *
FROM {CATALOG_NAME}.{SCHEMA_NAME}.gold_daily_summary
""")

spark.sql(f"""
CREATE OR REPLACE VIEW {CATALOG_NAME}.{SCHEMA_NAME}.vw_branch_summary AS
SELECT *
FROM {CATALOG_NAME}.{SCHEMA_NAME}.gold_branch_summary
""")

spark.sql(f"""
CREATE OR REPLACE VIEW {CATALOG_NAME}.{SCHEMA_NAME}.vw_customer_summary AS
SELECT *
FROM {CATALOG_NAME}.{SCHEMA_NAME}.gold_customer_summary
""")

spark.sql(f"""
CREATE OR REPLACE VIEW {CATALOG_NAME}.{SCHEMA_NAME}.vw_fraud_transactions AS
SELECT *
FROM {CATALOG_NAME}.{SCHEMA_NAME}.gold_fraud_transactions
""")

print("Views created successfully.")
```

```python
daily_df = spark.sql(f"""
SELECT
    transaction_date,
    total_transactions,
    total_amount,
    average_amount
FROM {CATALOG_NAME}.{SCHEMA_NAME}.vw_daily_summary
ORDER BY transaction_date
""")

display(daily_df)
```

```python
branch_df = spark.sql(f"""
SELECT
    branch,
    total_transactions,
    total_amount
FROM {CATALOG_NAME}.{SCHEMA_NAME}.vw_branch_summary
ORDER BY total_amount DESC
LIMIT 10
""")

display(branch_df)
```

```python
customer_df = spark.sql(f"""
SELECT
    customer_id,
    transaction_count,
    total_amount,
    highest_transaction
FROM {CATALOG_NAME}.{SCHEMA_NAME}.vw_customer_summary
ORDER BY total_amount DESC
LIMIT 20
""")

display(customer_df)
```

```python
fraud_df = spark.sql(f"""
SELECT
    fraud_reason,
    COUNT(*) AS fraud_count
FROM {CATALOG_NAME}.{SCHEMA_NAME}.vw_fraud_transactions
GROUP BY fraud_reason
ORDER BY fraud_count DESC
""")

display(fraud_df)
```
