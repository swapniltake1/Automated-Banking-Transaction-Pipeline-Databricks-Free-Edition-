```python
from datetime import datetime

PIPELINE = [
    ("00_Project_Setup", True),
    ("01_Precheck", True),
    ("02_Master_Data_Load", True),
    ("03_Ingestion", True),
    ("04_Data_Validation", True),
    ("05_Bronze_to_Silver", True),
    ("06_Fraud_Detection", False),
    ("07_Gold_Layer", True),
    ("08_Audit_Logging", False),
    ("09_SQL_Analytics", False),
    ("10_Dashboard", False)
]

print("="*90)
print("BANKING DATA ENGINEERING PLATFORM")
print("MASTER PIPELINE")
print("="*90)

start = datetime.now()

success = []
failed = []
```

```python
for notebook, critical in PIPELINE:

    print("="*90)
    print(f"Running : {notebook}")
    print("="*90)

    notebook_start = datetime.now()

    try:

        dbutils.notebook.run(notebook, timeout_seconds=0)

        duration = datetime.now() - notebook_start

        success.append(notebook)

        print(f"SUCCESS : {notebook}")
        print(f"Duration : {duration}")

    except Exception as ex:

        duration = datetime.now() - notebook_start

        failed.append(notebook)

        print(f"FAILED : {notebook}")
        print(str(ex))
        print(f"Duration : {duration}")

        if critical:
            raise Exception(
                f"Critical notebook failed : {notebook}. Pipeline stopped."
            )

        print("Continuing because notebook is non-critical...")
```

```python
end = datetime.now()

print("="*90)
print("PIPELINE EXECUTION SUMMARY")
print("="*90)
print(f"Start Time          : {start}")
print(f"End Time            : {end}")
print(f"Total Duration      : {end-start}")
print(f"Successful Notebooks: {len(success)}")
print(f"Failed Notebooks    : {len(failed)}")

print("\nSUCCESS LIST")
for n in success:
    print(f"  ✓ {n}")

if failed:
    print("\nFAILED LIST")
    for n in failed:
        print(f"  ✗ {n}")

print("="*90)

if failed:
    print("PIPELINE COMPLETED WITH WARNINGS")
else:
    print("PIPELINE COMPLETED SUCCESSFULLY")

print("="*90)
```
